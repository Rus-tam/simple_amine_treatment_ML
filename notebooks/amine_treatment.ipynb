```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn
import pickle
```

```python
data = pd.read_csv('../data/fixed_headers.csv', encoding = "cp1252", on_bad_lines='skip')
data
```

```python
data['temperature_inlet_liquid']
```

```python
clean_data = data.select_dtypes(include=['int', 'float'])
clean_data = clean_data.dropna()
clean_data
```

```python
clean_data.columns.to_list()
```

## Plot for H2S fraction in feed gas stream

```python
# Set the style
plt.style.use('dark_background')

# Define fig
fig, ax = plt.subplots(figsize=(15, 10))

# Plot the data
scatter = ax.scatter(x=np.arange(1, len(clean_data['h2s_mass_frac_inlet_vapour']) + 1),
                    y=clean_data['h2s_mass_frac_inlet_vapour'])

# Customize the plot
ax.set(title='Концентрация сероводорода в сырьевом газе',
      xlabel='Количество проб',
      ylabel='Концентрация сероводорода в сырьевом газе')

ax.plot();
fig.savefig('../spreadsheets/H2S_concentration.jpg')
```

## Machine learning model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

mass_frac_pred = clean_data[['mass_flow_inlet_vapour',
 'h2s_mass_frac_inlet_vapour', 'mass_flow_inlet_liquid',
 'h2o_mass_frac_inlet_liquid',
 'mea_mass_frac_inlet_liquid',
 'h2s_mass_frac_outlet_vapour']]

X = mass_frac_pred.drop('h2s_mass_frac_outlet_vapour', axis=1)
y = mass_frac_pred['h2s_mass_frac_outlet_vapour']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
score = model.score(X_test, y_test)
score
```

## Model examination

```python
vapour_mass_flow = [10000]
h2s_mass_frac = [0.1]
amine_mass_flow = [30000]
amine_h2o_mass_fr = [0.7]
amine_amine_mass_fr = [0.3]

examination_data = pd.DataFrame({'mass_flow_inlet_vapour': vapour_mass_flow, 'h2s_mass_frac_inlet_vapour': h2s_mass_frac,
                    'mass_flow_inlet_liquid': amine_mass_flow, 'h2o_mass_frac_inlet_liquid': amine_h2o_mass_fr,
                   'mea_mass_frac_inlet_liquid': amine_amine_mass_fr})

print(f'Примерное содержание сероводорода в очищенном газе составляет {model.predict(examination_data)[0]}')
```
